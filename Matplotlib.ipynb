```python
x = np.arange(0, 4* 3.141592653589793, 0.1)
y = np.sin(x)
plt.plot(x, y, 'mo')
plt.title("Wave Diagram", size=30, color='g')
plt.xlabel('X axis', size=20, color='r')
plt.ylabel('Y axis', size=20, color='r')
```

```python
x = np.arange(0, 5* np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1, 1)
plt.plot(x, y_sin, 'y')
plt.title('Sine')

plt.subplot(2,1, 2)
plt.plot(x, y_cos, 'r')
plt.title('Cosine')
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [250, 150, 210, 190]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.08, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True, startangle=300)

plt.axis('equal')
plt.show()
```

```python
# creating the dataset
data = {'C':20, 'C++':15, 'Java':30,'Python':35}
courses = data.keys()
values = data.values()
  
plt.figure(figsize = (10,5))
 
# creating the bar plot
plt.bar(courses, values, color ='cyan',width = 0.4)
 
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

```python
import matplotlib.pyplot as plt

data = {"Orange" : 20, "Guava": 40, "Mango" :60}
x = data.keys()
y = data.values()
plt.xlabel("Fruits", color="magenta", size='15')
plt.ylabel("Quantity sold", color="magenta", size='15')
plt.title("Food Distribution", color="yellow", size='20')
plt.bar(x, y)
plt.show()
```

```python
days = np.arange(1,8)
max_temp = np.random.randint(40,60, size=7)
min_temp = np.random.randint(20, 40, size=7)
avg_t = np.random.randint(35, 55, size=7)

print(days)
print(min_temp)
print(max_temp)
print(avg_t)

plt.figure(figsize=(15,7))
plt.plot(days)
plt.plot(min_temp, label= 'Min')
plt.plot(max_temp, label= 'Max')
plt.plot(avg_t, label= 'Avg')
plt.legend(loc='best', fontsize='large')
plt.show()
```
