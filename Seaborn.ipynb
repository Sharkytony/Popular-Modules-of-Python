```python
import matplotlib.pyplot as plt
sns.countplot(x=df['day'])
plt.xlabel('Days', size=15)
plt.xticks(fontsize='x-large')
plt.ylabel('Num. of customers', size=15)
plt.yticks(fontsize='x-large')
plt.title('Customers per day', size=15)
# plt.plot(df['tip'])
plt.show()
```

```python
df1 = sns.load_dataset('iris')
```

```python
df1.head()
```

```python
df1.corr()
```

```python
sns.heatmap(df1.corr())
```

Univariate analysis

```python
import numpy as np
df1_setosa = df1.loc[df1['species'] == 'setosa']
df1_virginica = df1.loc[df1['species'] == 'virginica']
df1_versicolor = df1.loc[df1['species'] == 'versicolor']
np.zeros_like
```

```python
import matplotlib.pyplot as plt
plt.scatter(df1_setosa['sepal_length'], np.zeros_like(df1_setosa['sepal_length']))
plt.scatter(df1_virginica['sepal_length'], np.zeros_like(df1_virginica['sepal_length']))
plt.scatter(df1_versicolor['sepal_length'],np.zeros_like(df1_versicolor['sepal_length']))
plt.xlabel('Petal length')
plt.show()
```

Bivarite analysis

```python
sns.catplot(x='sepal_length',y='sepal_width',hue='species',palette='CMRmap', data=df1)
```

```python
sns.barplot(x='petal_length',y='petal_width',hue='species',data=df1,palette='Accent')
```

```python
sns.displot(df1['sepal_length'], kde=True, palette='Accent_r', bins=15)
```

```python
sns.histplot(data=df1, x='sepal_width', bins=5, palette='RdBu')
```

```python
sns.boxplot(data=df1)
```

```python
sns.displot(data=df1, x='petal_length', y='sepal_length', hue='species')
```

```python
sns.catplot(x='sepal_length', y='petal_width', data= df1, hue='species') #works same as catplot
plt.xticks(rotation=90, fontsize="xx-small")
plt.show()
```

Multivariate analysis

```python
sns.jointplot(x='sepal_width', y='petal_length', hue='species', data=df1, kind='scatter', color='m')
```

```python
sns.kdeplot(x='sepal_width',y='petal_width',data=df1, )
```

```python
sns.countplot(x='species', data=df1)
```

```python
sns.pairplot(data=df1, hue='species')
```
